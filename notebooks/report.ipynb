# Summary

I selected Sex, Pclass, and Age as the main factors for estimating survival, and also considered the number of companions (family size) as a supplementary variable.



# Sex

It is clear that women were far more likely to survive than men.
More than two-thirds of female passengers were rescued, while the survival rate of male passengers was below 20%.
Although there were more men than women on board, the number of female survivors was substantially higher.



# Pclass

This figure shows that passengers in the first class were much more likely to survive than those in the second or third class.
In particular, a large proportion of the victims belonged to the third class.
It is likely that upper-class passengers were prioritized for evacuation or that their cabins were located in areas with easier access to lifeboats.




This tendency can also be interpreted through the positive correlation between fare and survival.
The higher the ticket price, the greater the likelihood of survival, which supports the pattern observed in the Pclass analysis.

# Age

It is notable that in the youngest group (ages 05), the number of survivors was significantly higher than that of non-survivors.
A similar pattern appears in other young age groups (515), where the survival rate was higher than among adults.
These findings suggest that children were given priority during evacuation.



# Family size

Passengers traveling with family members appeared slightly more likely to survive, but this result is not entirely reliable.
Most passengers were traveling alone, and family groups often included children, who were likely to receive priority assistance.
